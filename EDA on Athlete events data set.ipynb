### Finding insights from the data 

#### feature engineering using the column medal

```python
df.Medal.unique()
```

```python
##### Lets create seperate column for each medal
df['Gold']=[1 if medal=='Gold' else 0 for medal in df['Medal']]
df['Silver']=[1 if medal=='Silver' else 0 for medal in df['Medal']]
df['Bronze']=[1 if medal=='Bronze' else 0 for medal in df['Medal']]
```

```python
df.head(5)
```

### 1.Which country won most medal

```python
country_wise_medal=df.groupby(['Team'])['Gold', 'Silver','Bronze'].sum()
```

```python
country_wise_medal
```

```python
#Adding a column  Total to the object country_wise_medal
```

```python
country_wise_medal['Total']=country_wise_medal['Gold']+country_wise_medal['Silver']+country_wise_medal['Bronze']
```

```python
country_wise_medal
```

```python
country_wise_medal.sort_values(by='Total',ascending=False)
```

```python
country_wise_medal.nlargest(5,'Total')
```

#### :America got most no of medal

### 2.Country with no gold medal

```python
no_gold_medal=country_wise_medal[country_wise_medal["Gold"]==0]
no_gold_medal
```

#### :There are 942 countries with no gold medal

### 3.Is India included in the above list

```python
India_medal=df.query("(Gold==1 or Silver==1 or Bronze==1 )& NOC=='IND'")
India_medal
```

#### :India have got gold medals in Hockey,so its not in the no_gold_medal countries list
