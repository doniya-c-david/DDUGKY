```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./titanic.csv')
data.head()
```

```python
data.drop(columns=['PassengerId','Ticket','Cabin'],inplace=True)
```

```python
data.sample()
```

#### Groupby

 Gender wise survival data

```python
sex_survived=data.groupby(['Sex']) ['Survived'].sum()
```

```python
sex_survived
```

```python
sex_Sibsp=data.groupby(['Sex']) ['SibSp'].sum()
sex_Sibsp
```

Gender wise count  of each  passenger class

```python
cls_dat=data.groupby(['Sex','Pclass'])['Pclass'].sum()
cls_dat
```

#### Quering the data

```python
sex_age_18= data.query("Age<=22.0  & Sex=='female'")
```

```python
sex_age_18
```

```python
male_age_under_20=data.query("Age<20  & Sex=='male'")
male_age_under_20
```

#### Pivot table

```python
pivot_01=data.pivot_table(values='Fare',index='Sex',columns='Pclass',aggfunc='sum')
pivot_01
```

```python
pivot_02=data.pivot_table(values='Age',index='Sex',columns='Embarked',aggfunc='mean')
pivot_02
```

#### pd.cut

```python
cutoff=[0,18,50,85]
labels=['Child','Adult','Old']
data["Age_type"]=pd.cut(data.Age,bins=cutoff,labels=labels)
```
