# Plotting with pandas

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Car_sales.csv')
```

```python
data.sample(10)
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.rename(columns={'__year_resale_value':'year_resale_value'},inplace=True)
```

```python
data.sample()
```

## Plots on numerical value columns

### Histogram

```python
data.Horsepower.plot.hist(figsize=(6,3), cmap='brg');
```

### Boxplot

```python
data.Sales_in_thousands.plot.box(figsize=(4,3),cmap='gist_earth')
```

```python
 data.Horsepower.plot.box(figsize=(4,3),cmap='Blues_r')  
```

### Line chart

```python
data.sort_values(by='Engine_size').plot.line('Engine_size','Horsepower', figsize=(6,3),cmap='winter');
```

### Barplot

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2),cmap='inferno');
```

```python
data.head()
```

```python
data[:20].plot.bar('Manufacturer','Sales_in_thousands',figsize=(6,3))
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3));
```

### Area Chart

```python
data.plot.area('Manufacturer','Sales_in_thousands',cmap='Pastel2');
```

### Horizontal Bars

```python
data[:10].plot.barh('Model','Power_perf_factor',fig=(6,4));
```

```python
data.sort_values(by='Sales_in_thousands',ascending=False)[:10].plot.barh('Model','Power_perf_factor',fig=(6,4),cmap="winter");
```
